```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_geo = pd.read_csv('./../../data/raw/table_jonction.csv' , sep=';')
df_geo.describe()

required_geo_cols = ["NatlStationCode", "Latitude", "Longitude"]
missing_geo_cols = [c for c in required_geo_cols if c not in df_geo.columns]
if missing_geo_cols:
    raise ValueError(f"table_jonction.csv missing columns: {missing_geo_cols}")

# On ne garde que les colonnes utiles du df1 (lat/lon + clé)
df_geo = df_geo[required_geo_cols].rename(
    columns={
        "NatlStationCode": "code site",
        "Latitude": "lat",
        "Longitude": "lon",
    }

)
df_geo

```
